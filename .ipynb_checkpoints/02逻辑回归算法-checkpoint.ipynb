```python
#代价函数
def costFunctionReg(theta,X,y,mylambda=1):
    #样本数量
    m=y.size
    #参数的拷贝
    tmp_theta=theta.reshape(X.shape[1],1).copy()
    #预测函数
    z=np.array(dot(X,tmp_theta))
    h=1.0/(1.0+np.e**-z)
    #第一个参数不计算，和L* θ 效果类似
    tmp_theta[θ ,θ ]=θ 
    #代价函数计算
    J=1.0/m*(-y.T.dot(log(h))-(1-y).T.dot(log(1-h)))+1.0*mylanbda/2/m*sum(tmp_theta**2)
    if np.isnan(J):
        return np.inf
    return J

#梯度
def gradientReg(theta,X,y,mylambda=1):
    #样本数量
    m=y.size
    #参数的拷贝
    tmp_theta=theta.reshape(X.shape[1],1).copy()
    #预测函数
    z=np.array(dot(X,tmp_theta))
    h=1.0/(1.0+np.e**-z)
    #第一个参数不计算，和L·θ 效果类似
    tmp_theta[θ ,θ ]=θ 
    #梯度计算
    grad=1.0/m*X.T.dot(h-y)+1.0*mylambda/m*tmp_theta
    #结果变成一行
    grad=grad.flatten()
    
    return grad
```

```python
import numpy as np
import matplotlib.pyplot as plt

#def loaddate
def loaddata(filename):
    file=open(filename)
    x=[]
    y=[]
    for line in file.readlines():
        line=line.strip().split()
        x.append([1,float(line[0]),float(line[1])])
        y.append(float(line[-1]))
    xmat=np.mat(x)
    ymat=np.mat(y).T
    file.close()
    return xmat,ymat
# w calc
def w_cal(xmat,ymat,alpha=0.001,maxIter=10001):
    W=np.mat(np.random.randn(3,1))
    w_save=[]
    for i in range(maxIter):
        H=1/(1+np.exp(-xmat*W))
        dw=xmat.T*(H-ymat)
        W-=alpha*dw
        if i%100==0:
            w_save.append([W.copy(),i])
    return W,w_save
#implement
xmat,ymat=loaddata('ytb_lr.txt')
print(xmat.shape)
print(ymat.shape)
print(xmat)
print(ymat)
```

```python
W,w_save = w_cal(xmat,ymat)
print(W)
print('xmat:',xmat)
print('ymat:',ymat)
 
# 基本展示
w0 = W[0,0]
w1 = W[1,0]
w2 = W[2,0]
plotx1 =  np.arange(1,7,0.001)
plotx2 = -w0 /w2 - w1/w2 * plotx1
plt.plot(plotx1,plotx2,c = 'r',label = 'decision boundary')

plt.scatter(xmat[:,1][ymat ==0].A,xmat[:,2][ymat ==0].A,marker='^',s = 150,label = 0)
plt.scatter(xmat[:,1][ymat ==1].A,xmat[:,2][ymat ==1].A,s = 150,label = 1)
plt.grid()
plt.legend()
plt.show()
```

```python
for wi in w_save:
    plt.clf()
    w0 = wi[0][0,0]
    w1 = wi[0][1,0]
    w2 = wi[0][2,0]
    plotx1 =  np.arange(2,6,0.01)
    plotx2 = -w0 /w2 - w1/w2 * plotx1
    plt.plot(plotx1,plotx2,c = 'r',label = 'decision boundary')
 
    plt.scatter(xmat[:,1][ymat ==0].A,xmat[:,2][ymat ==0].A,marker='^',s = 150,label = 0)
    plt.scatter(xmat[:,1][ymat ==1].A,xmat[:,2][ymat ==1].A,s = 150,label = 1)
    plt.grid()
    plt.legend()
    plt.title('iter:%s'%np.str(wi[1]))
    plt.pause(0.001)
plt.show()
```
